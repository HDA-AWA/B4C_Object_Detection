```python
from ultralytics import YOLO

model = YOLO('C:/Users/Admin/Documents/datasetTesting/runs/detect/elevator/weights/best.pt')

CONFIDENCE_THRESHOLD = 0.8  

results = model('door.jpeg')

for result in results:
    detected = False  

    for box, label, score in zip(result.boxes.xyxy, result.boxes.cls, result.boxes.conf):
        if score >= CONFIDENCE_THRESHOLD:
            detected = True
            x1, y1, x2, y2 = map(int, box)
            label_text = result.names[int(label)]
            print(f"Detected: {label_text} | Confidence: {score:.2f} | Box: ({x1}, {y1}), ({x2}, {y2})")

    if detected:
        result.show()
```

```python
import cv2

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

while True:
    ret, frame = cap.read()

    if ret:
        results = model(frame)
        
        for result in results:
            boxes = result.boxes.xyxy
            labels = result.boxes.cls
            scores = result.boxes.conf

            for box, label, score in zip(boxes, labels, scores):
                x1, y1, x2, y2 = map(int, box)
                label_text = result.names[int(label)]
                confidence = score

                cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 255, 0), 3)

                text = f"{label_text}: {confidence:.2f}"
                (text_width, text_height), baseline = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, 0.8, 2)

                cv2.rectangle(frame, (x1, y1 - text_height - 10), (x1 + text_width, y1), (255, 255, 0), -1)
                cv2.putText(frame, text, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0), 2)

        cv2.imshow('Webcam Output', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

while True:
    ret, frame = cap.read()

    if ret:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        blur = cv2.GaussianBlur(gray, (5, 5), 0)
        edges = cv2.Canny(blur, 50, 150) 
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5)) 
        edges = cv2.dilate(edges, kernel, iterations=1) 

        edges_colored = cv2.merge((edges, np.zeros_like(edges), np.zeros_like(edges)))

        frame_with_edges = cv2.addWeighted(frame, 0.9, edges_colored, 0.5, 0)

        results = model(frame)
        
        for result in results:
            boxes = result.boxes.xyxy
            labels = result.boxes.cls
            scores = result.boxes.conf

            for box, label, score in zip(boxes, labels, scores):
                x1, y1, x2, y2 = map(int, box)
                label_text = result.names[int(label)]
                confidence = score

                cv2.rectangle(frame_with_edges, (x1, y1), (x2, y2), (255, 255, 0), 3)

                text = f"{label_text}: {confidence:.2f}"
                (text_width, text_height), baseline = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, 0.8, 2)

                cv2.rectangle(frame_with_edges, (x1, y1 - text_height - 10), (x1 + text_width, y1), (255, 255, 0), -1)
                cv2.putText(frame_with_edges, text, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0), 2)

        cv2.imshow('Edge Detection + Object Detection', frame_with_edges)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
#live detection

import cv2
from ultralytics import YOLO

model = YOLO('C:/Users/Admin/Documents/onlydoors/runs/detect/train/weights/best.pt')

CONFIDENCE_THRESHOLD = 0.4  

video_path = 'dogvid1.mp4'
output_path = 'C:/Users/Admin/Documents/datasetTesting/outputVid/output_video.mp4'

cap = cv2.VideoCapture(video_path)

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = int(cap.get(cv2.CAP_PROP_FPS))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

while True:
    ret, frame = cap.read()

    if not ret:
        break  

    results = model(frame)

    for result in results:
        boxes = result.boxes.xyxy
        labels = result.boxes.cls
        scores = result.boxes.conf

        for box, label, score in zip(boxes, labels, scores):
            if score < CONFIDENCE_THRESHOLD:
                continue

            x1, y1, x2, y2 = map(int, box)
            label_text = result.names[int(label)]
            confidence = score

            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 255, 0), 3)

            text = f"{label_text}: {confidence:.2f}"
            (text_width, text_height), baseline = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, 0.8, 2)

            cv2.rectangle(frame, (x1, y1 - text_height - 10), (x1 + text_width, y1), (255, 255, 0), -1)
            cv2.putText(frame, text, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0), 2)

    out.write(frame)
    cv2.imshow('Video Output', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
